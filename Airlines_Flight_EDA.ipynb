#### Q.1. What are the airlines in the dataset, accompanied by their frequencies?

```python
# showing all the airlines with their frequency
df['airline'].value_counts().reset_index()
```

```python
# showing the airlines with their frequecy in a horizantal bar plot

airline1 = df['airline'].value_counts().reset_index()

plt.figure(figsize=(14,4.5))
barplt = sns.barplot(
        y = 'airline', 
        x='count', 
        data=airline1,
        color='skyblue')

for bars in barplt.containers:
    barplt.bar_label(bars, padding=3)

plt.title('Airlines With There Frequency', fontsize=20)
plt.xlabel('Frequency')
plt.tight_layout()
```

#### Q.2. Show Bar graph representing the Departure Time and Arrival Time

```python
# showing the Arrival time for the flight

df['arrival_time'].value_counts().reset_index()
```

#### **Observation:**

-> departure time:

* most flights depart in Morning (71000+), followed by early_morning (66000+) and evening (65000+)

* very few flights leave in late_night (1300)

-> arrival time:

* most flights arrive in night (91000+) and evening (78000+)

* very less flights arrive in early_morning (15000) and late_night (14000)

#### Q.4. Does price varies with airlines ?

```python
df.groupby(['airline','class'], as_index=False)['price'].mean()
```

```python
# Drawing a category plot for the airlines and their mean price in every class

plt.figure(figsize=(12, 4))
sns.catplot(
    x='airline',
    y='price',
    data=df,
    hue='class',
    kind='bar',
    palette='rocket')
plt.title('Mean Price of Flights by Airline and Class')
plt.show()
```

#### Observations:

Yes, price significantly varies by airline and class:
Business class tickets for Vistara and Air India are the most expensive, averaging around 55,000.
Economy class prices remain much lower, with AirAsia offering the most affordable fares.
Budget airlines like GO_FIRST, Indigo, and SpiceJet maintain consistent and lower ticket prices.

#### Q.5. Which airline offers the cheapest average price for Economy class flights?

```python
df[df['class'] == 'Economy'].groupby('airline', as_index=False)['price'].mean().sort_values(by='price')
```

```python
# Calculate average price of Economy class by airline

economy_avg = df[df['class'] == 'Economy'].groupby('airline')['price'].mean().sort_values()

plt.figure(figsize=(15,3.5))
plot1 = sns.barplot(
    y = economy_avg.index,
    x = economy_avg.values,
    color="orange"
    )
plt.title('Average Price of Economy class Flight by Airlines', fontsize=14 , fontweight='bold')
plt.xlabel('Average price')
plt.tight_layout()

for bars in plot1.containers:
    plot1.bar_label(bars, padding=5, fontweight='bold')
```

#### Observation

* AirAsia offers the lowest average price for Economy class flights, making it the most budget-friendly option.
* Indigo and GO_FIRST are slightly higher but still in the affordable range.
* SpiceJet and Air India show mid-range pricing.
* Vistara has the highest average price among all airlines for Economy class.

#### Insight:
Passengers looking for economical tickets should prefer AirAsia or Indigo, while those choosing Vistara should expect premium pricing even in Economy class.

#### Q.7. Which source–destination city pairs have the highest and lowest ticket prices?

```python
route_price = df.groupby(['source_city','destination_city'])['price'].mean().sort_values(ascending=False)


# Plot top 10 expensive and top 10 cheapest routes
plt.figure(figsize=(18,5))
route_price[:10].plot(kind='bar', color='green', label='Top 10 Cheapest')
plt.title('Top 10 highest Routes by Average Price', fontsize=20, fontweight='bold')
plt.xlabel('Route')
plt.ylabel('avg_price')
plt.legend()
plt.xticks(rotation=75)

plt.figure(figsize=(18,5))
route_price[-10:].plot(kind='bar', color='red', label='Top 10 Cheapest')
plt.title('Top 10 Cheapest Routes by Average Price', fontsize=20, fontweight='bold')
plt.xlabel('Route')
plt.ylabel('avg_price')
plt.legend()
plt.xticks(rotation=75)

plt.show()
```

#### Observation

#### Top 10 Expensive Routes:

* The most expensive route is Chennai to Bangalore.
* The top expensive routes are primarily between major business hubs, suggesting high demand for these city pairs.

#### Top 10 Cheapest Routes:
                                                                                            
* The cheapest route is Hyderabad to Mumbai.
* Cheaper routes also connect major cities, but at significantly lower average prices.

#### Q.8. Does ticket price change based on the departure time and arrival time?

```python
df.head()
```

```python
# checking the mean price based on departure_time
df.groupby('departure_time', as_index=False)['price'].mean().sort_values(by='price' ,ascending=False)
```

```python
# checking the mean price based on arrival time

df.groupby('arrival_time', as_index=False)['price'].mean().sort_values(by='price', ascending=False)
```

```python
# Showing a category graph for departure and arrival time with ticket price

plt.figure(figsize=(19, 4.5))
plt.subplot(1,2,1)
# plt.tight_layout
plt1 = df.groupby('departure_time', as_index=False)['price'].mean()
plt11 = sns.barplot(
    x='departure_time',
    y = 'price', 
    data=plt1, 
    hue='departure_time',
    palette='Blues',
    legend=False)
plt.title('Departure Time VS Ticket Price', fontsize=18, fontweight='bold')

for bars in plt11.containers:
    plt11.bar_label(bars)

plt.subplot(1,2,2)

plt2 = df.groupby('arrival_time', as_index=False)['price'].mean()
plt22 = sns.barplot(
    x='arrival_time',
    y='price',
    data=plt2,
    hue='arrival_time',
    palette='Greens',
    legend=False)
plt.title('Arrival Time VS Ticket Price', fontsize=18, fontweight='bold')

for bars in plt22.containers:
    plt22.bar_label(bars)

plt.show()
```

#### Observation

Departure Time
* Flights departing at Night have the highest Average ticket price
* Flights departing at Late_Night have the lowest Average ticket Price
* Evening and Morning departures also have relatively higher prices compared to Afternoon.

Arrival Time
* Flights arriving at Evening have the highest Average ticket Price
* Flights arriving at Late_Night have the lowest Average ticket Price
* Morning and Night arrival time also have relatively higher prices compared to Afternoon and Early_morning

#### Q.10. How is the price affected when tickets are bought in just 1 or 2 days before departure?

```python
# checking the mean ticket price for differents days left

df.groupby('days_left')['price'].mean()
```

```python
# Showing the ticket price with days left

sns.relplot(x='days_left', y='price', kind='line', data=df)

plt.title('ticket price with days left')
plt.show()
```

#### Observation

* Ticket price is very high when booked very close to departure date (1-2 days left) , average price around 30,000+
* Prices start dropping steadily after 10 days left, stabilizing around 18,000–20,000 when booked 20+ days in advance.
* The clear downward trend shows that last-minute bookings are significantly more expensive, whereas early bookings result in much cheaper fares.
